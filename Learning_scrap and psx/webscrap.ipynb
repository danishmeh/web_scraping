```python
url = "https://dps.psx.com.pk/listings?page=1"
r = requests.get(url)
print(r.text)
```

```python
soup = BeautifulSoup(r.text,"lxml")
print(soup)
```

```python
response = requests.get("https://dps.psx.com.pk/listings").text
```

```python
soup = BeautifulSoup(response,'lxml')
```

```python
print(soup.prettify())
```

```python
soup.find_all('p')
```

```python
# Find all the <div> elements with class="pageHeader__title"
div_elements = soup.find_all('div', class_="header__mobile")

# Iterate through the list of <div> elements and extract their text
for div in div_elements:
    print(div.text.strip())
```

```python
# Find all the <div> elements with class="pageHeader__title"
div_elements = soup.find_all('div', class_="topIndices__item")

# Iterate through the list of <div> elements and extract their text
for div in div_elements:
    print(div.text.strip())
```
