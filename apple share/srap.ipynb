```python
import pandas as pd
import numpy as np
import requests
from bs4 import BeautifulSoup
```

```python
url ="https://en.wikipedia.org/wiki/World_population"
response = requests.get(url)
Soup = BeautifulSoup(response.content,"html.parser")
```

```python
Soup.text
```

```python
Soup.title
```

```python
print(Soup.title.text)
```

```python
# print(Soup.prettify())
# How to Find All Table

tables = Soup.find_all("table")
# print(tables)
```

```python
dataframe = []
##The code then iterates over each table using the enumerate() function to get both the index (i) and the table 
# itself (table).
for i, table in enumerate(tables):
    #For each table, it finds all the rows (<tr>) within the table using the find_all("tr") 
    # method. The [1:] indexing is used to exclude the header row (assumed to be the first row).
    rows = table.find_all("tr")[1:]
    data = [] # It initializes an empty list called data to store the data from each row
    for row in rows:
        cols = row.find_all("td")
        cols = [col.text.strip() for col in cols]
        data.append(cols)
    df = pd.DataFrame(data)
    dataframe.append(df) 
           
```
